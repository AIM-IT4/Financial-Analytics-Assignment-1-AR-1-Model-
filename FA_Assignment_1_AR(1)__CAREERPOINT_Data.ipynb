```python
Exponential_smoothing = create_model('exp_smooth')
```

## Creating ARIMA model

```python
arima = create_model('arima',order=(1,0,0))
```

```python
tuned_arima = tune_model(arima)
```

```python
def summarize_model(model):
    """
    Provides statistical summary for some statistical models
    """
    # Statistical Summary Table
    try:
        print(model.summary())
    except:
        print("Summary does not exist for this model.")
```

```python
summarize_model(tuned_arima)
```

```python
def get_residual_properties(model):
    """
    Plots and displays variance of residuals
    """
    #### Residuals ----
    try:
        plot_data = exp_name.plot_model(
            model,
            plot="residuals",
            return_data=True
        )
        resid = plot_data['data']
        resid_sigma2 = (resid**2).sum()/(len(resid))
        print(f"Residual Sigma2: {resid_sigma2}")
    except:
        print("Residuals can not be extracted for this model.")
```

```python
get_residual_properties(tuned_arima)
```
